Container types in Python
Python provides several built-in container types used to store and organize collections of other objects. These are broadly categorized into four main types: lists, tuples, dictionaries, and sets.
Built-in Container Types
Container 	Description	Mutability	Order	Key/Index Access
List	A sequence of items, which can be of different types.	Mutable (can be changed after creation)	Ordered (elements maintain their defined order)	Indexed by integers
Tuple	An immutable sequence of items, similar to a list but cannot be modified after creation.	Immutable	Ordered	Indexed by integers
Dictionary	A collection of key-value pairs, where each unique key is associated with a value.	Mutable	Ordered (since Python 3.7)	Accessed by keys
Set	An unordered collection of unique elements, useful for membership testing and mathematical set operations (union, intersection, etc.).	Mutable	Unordered (cannot use index/key access)	No
String	A sequence of characters.	Immutable	Ordered	Indexed by integers

```python
# Problem 1: Confusion Matrix Metrics
# Binary classification model for flood prediction

# Given values
TP = 50  # True Positives
FP = 8   # False Positives
TN = 32  # True Negatives
FN = 10  # False Negatives

# Task 1: Calculate total predictions
total_predictions = TP + FP + TN + FN
print("Total Predictions:", total_predictions)

# Task 2: Compute metrics
# Accuracy = (TP + TN) / (TP + FP + TN + FN)
accuracy = (TP + TN) / (TP + FP + TN + FN)
print("Accuracy:", accuracy)

# Precision = TP / (TP + FP)
precision = TP / (TP + FP)
print("Precision:", precision)

# Recall = TP / (TP + FN)
recall = TP / (TP + FN)
print("Recall:", recall)

# F1 Score = 2 * (Precision * Recall) / (Precision + Recall)
F1 = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", F1)
```

```python
# Problem 2: R² (Coefficient of Determination)
# Regression model predicting river discharge

# Given values
SSR = 420  # Sum of Squares due to Regression
SSE = 180  # Sum of Squares due to Error

# Task 1: Compute Total Sum of Squares (SST)
# SST = SSR + SSE
SST = SSR + SSE
print("Total Sum of Squares (SST):", SST)

# Task 2: Calculate R²
# R² = SSR / SST
R_squared = SSR / SST
print("R² (R-squared):", R_squared)

# Task 3: Interpret whether the model explains more than 70% variance
if R_squared > 0.70:
    print("Interpretation: The model explains more than 70% of the variance.")
else:
    print("Interpretation: The model explains less than or equal to 70% of the variance.")
```

```python
# Problem 3: Mean Squared Error (MSE)
# Temperature prediction model

# Task 1: Store squared errors in Python
squared_errors = [2, 4, 1, 9, 4]  # Squared errors for 5 observations
print("Squared errors:", squared_errors)

# Task 2: Calculate Mean Squared Error (MSE)
# MSE = sum of squared errors / number of observations
n = 5  # Number of observations
sum_squared_errors = 2 + 4 + 1 + 9 + 4
MSE = sum_squared_errors / n

# Task 3: Print the result
print("Mean Squared Error (MSE):", MSE)
```

```python
# Problem 4: Cook's Distance - Influential Observation
# Regression model analysis

# Given values
e_squared = 9  # Squared residual
MSE = 3        # Mean Squared Error
p = 3          # Number of predictors
h = 0.2        # Leverage value
n = 25         # Sample size

# Task 1: Compute Cook's Distance
# Formula: D_i = (e_i^2 / (p * MSE)) * (h_ii / (1 - h_ii)^2)
D_i = (e_squared / (p * MSE)) * (h / ((1 - h) ** 2))
print("Cook's Distance (D_i):", D_i)

# Task 2: Compare with threshold 4/n
threshold = 4 / n
print("Threshold (4/n):", threshold)

# Task 3: Decide whether the observation is influential
if D_i > threshold:
    print("Decision: The observation is INFLUENTIAL (D_i > 4/n)")
else:
    print("Decision: The observation is NOT INFLUENTIAL (D_i <= 4/n)")
```

```python
# Problem 5: Simple Linear Regression Prediction (Calculator Style)
# Crop yield prediction model

# Given regression model: y = 2.5x + 10
# Given rainfall value
x = 12  # Rainfall value

# Task 1: Predict the value of y
# Using the formula: y = 2.5x + 10
y = 2.5 * x + 10
print("Predicted crop yield (y):", y)

# Task 2: Print the predicted output (already done above)

# Task 3: Explain the meaning of slope and intercept
print("\nExplanation:")
print("Slope (2.5): For every unit increase in rainfall (x), the crop yield increases by 2.5 units.")
print("Intercept (10): When rainfall is zero (x=0), the baseline crop yield is 10 units.")
```
