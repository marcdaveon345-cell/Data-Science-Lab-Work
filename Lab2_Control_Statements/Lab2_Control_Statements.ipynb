```python
Main_list = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Create a new list containing only the first element of each sublist
first_elements = [row[0] for row in matrix]
print("First elements of each row:", first_elements)

# Flatten the list of lists into a single list
flattened_list = [element for row in matrix for element in row]
print("Flattened list:", flattened_list)

# Create a new matrix where each element is doubled
doubled_matrix = [[element * 2 for element in row] for row in matrix]
print("Doubled matrix:", doubled_matrix)

# Extract elements greater than 5 from the flattened list
elements_greater_than_5 = [element for row in matrix for element in row if element > 5]
print("Elements greater than 5:", elements_greater_than_5)
```
